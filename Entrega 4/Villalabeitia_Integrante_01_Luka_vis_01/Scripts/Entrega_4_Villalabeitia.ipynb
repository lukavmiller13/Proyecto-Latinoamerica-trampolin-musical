```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
canciones = ["Ando - Jere Klein", "Perro Negro - Bad Bunny / Feid", "¿Cuál es su nombre? - Lucky Browne / Jere Klein / Mateo on the Beatz",
            "X ESO BB - Jere Klein / Nicki Nicole", "Lollipop - Darell"]

semanas = ["Semana Antepasada", "Semana Pasada", "Esta Semana"]

reproducciones = np.array([
    [3843640, 2458095, 2440258, 2324929, 1416842],
    [3846585, 2571480, 2378428, 2226384, 2060024],
    [3418306, 2163924, 2090600, 1993795, 1767105]
])

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))

for i, cancion in enumerate(canciones):
    plt.plot(semanas, reproducciones[:, i], marker='o', label=cancion)

# Configurar el gráfico
plt.title('Top 5 Canciones en Chile - Reproducciones')
plt.xlabel('Semanas')
plt.ylabel('Reproducciones (en millones)')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
canciones = ["Ando - Jere Klein", "Perro Negro - Bad Bunny / Feid", "¿Cuál es su nombre? - Lucky Browne / Jere Klein / Mateo on the Beatz",
            "X ESO BB - Jere Klein / Nicki Nicole", "Lollipop - Darell"]

reproducciones_semana_actual = [3418306, 2163924, 2090600, 1993795, 1767105]
reproducciones_semana_pasada = [3846585, 2571480, 2378428, 2226384, 2060024]
reproducciones_semana_antepasada = [3843640, 2458095, 2440258, 2324929, 1416842]

# Crear el gráfico de líneas
fig, ax = plt.subplots(figsize=(10, 6))

# Barra para la semana antepasada
bar_width = 0.25
bar_positions = np.arange(len(canciones))

ax.barh(bar_positions - bar_width, reproducciones_semana_antepasada, height=bar_width, label='Semana Antepasada')
ax.barh(bar_positions, reproducciones_semana_pasada, height=bar_width, label='Semana Pasada')
ax.barh(bar_positions + bar_width, reproducciones_semana_actual, height=bar_width, label='Esta Semana')

# Configurar el gráfico
ax.set_yticks(bar_positions)
ax.set_yticklabels(canciones)
ax.set_xlabel('Reproducciones (en millones)')
ax.set_title('Top 5 Canciones en Chile - Reproducciones')
ax.legend()
ax.grid(axis='x')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Nombres actualizados de las canciones
canciones = ["Ando - Jere Klein", "Perro Negro - Bad Bunny / Feid", "¿Cuál es su nombre? - Lucky Browne / Jere Klein / Mateo on the Beatz",
            "X ESO BB - Jere Klein / Nicki Nicole", "Lollipop - Darell"]
esta_semana = [3418306, 2163924, 2090600, 1993795, 1767105]

# Crear un array de índices para la ubicación de las barras
ind = np.arange(len(canciones))

# Ajustar el ancho de la figura y el ancho de las barras
fig = plt.figure(figsize=(10, 6))
width = 0.7

# Crear el gráfico de barras
bars = plt.bar(ind, esta_semana, width, color='blue')

# Etiquetas, título y leyenda
plt.xlabel('Canciones')
plt.ylabel('Número de reproducciones')
plt.title('Top 5 Canciones en Chile - Esta Semana')
plt.xticks(ind, canciones, rotation=15, ha="right")

# Ajustar el límite superior del eje y para dejar espacio adicional
plt.ylim(0, max(esta_semana) * 1.1)

# Formatear el eje y para quitar los puntos decimales
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: format(int(x), ',')))

# Añadir las cifras exactas arriba de las barras
for bar, count in zip(bars, esta_semana):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.02 * max(esta_semana), f'{count:,}', ha='center', va='bottom')

# Ajustar el diseño para evitar recorte de etiquetas
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
