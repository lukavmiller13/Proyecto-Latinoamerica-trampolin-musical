```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
canciones = ["Que Onda - Calle 24 / Chino Pacas / Fuerza Regida", "HARLEY QUINN - Fuerza Regida / Marshmello", "LOU LOU - Gabito Ballesteros / Natanel Cano", "LADY GAGA - Peso Pluma / Gabito Ballesteros / Junior H", "ELOVGRA - Alex Favela / Grupo marca Registrada / Joaquín Medina"]
semanas = ["Semana Antepasada", "Semana Pasada", "Esta Semana"]

# Números de reproducciones en millones
reproducciones = [
    [11889528, 0, 0, 9766313, 9000768],
    [13184649, 3863270, 9258317, 9777236, 9003113],
    [14032655, 12915956, 11003022, 9780868, 8983008]
]

# Crear el gráfico de barras horizontales
plt.figure(figsize=(12, 8))

bar_height = 0.2
bar_positions = np.arange(len(canciones))

for i, semana in enumerate(semanas):
    plt.barh(bar_positions + i * bar_height, reproducciones[i], height=bar_height, label=semana)

# Configurar el gráfico
plt.title('Top 5 Canciones - México')
plt.ylabel('Canciones')
plt.xlabel('Número de reproducciones (en millones)')
plt.yticks(bar_positions + (len(semanas) - 1) * bar_height / 2, canciones)
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Nombres actualizados de las canciones
canciones = [
    "Que Onda - Calle 24 / Chino Pacas / Fuerza Regida",
    "HARLEY QUINN - Fuerza Regida / Marshmello",
    "LOU LOU - Gabito Ballesteros / Natanel Cano",
    "LADY GAGA - Peso Pluma / Gabito Ballesteros / Junior H",
    "ELOVGRA - Alex Favela / Grupo marca Registrada / Joaquín Medina"
]

esta_semana = [14032655, 12915956, 11003022, 9780868, 8983008]

# Crear un array de índices para la ubicación de las barras
ind = np.arange(len(canciones))

# Ajustar el ancho de la figura y el ancho de las barras
fig = plt.figure(figsize=(10, 6))
width = 0.7

# Crear el gráfico de barras
bars = plt.bar(ind, esta_semana, width, color='blue')

# Etiquetas, título y leyenda
plt.xlabel('Canciones')
plt.ylabel('Número de reproducciones')
plt.title('Top 5 Canciones en México - Esta Semana')
plt.xticks(ind, canciones, rotation=15, ha="right")

# Ajustar el límite superior del eje y para dejar espacio adicional
plt.ylim(0, max(esta_semana) * 1.1)

# Formatear el eje y para quitar los puntos decimales
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: format(int(x), ',')))

# Añadir las cifras exactas arriba de las barras
for bar, count in zip(bars, esta_semana):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.02 * max(esta_semana), f'{count:,}', ha='center', va='bottom')

# Ajustar el diseño para evitar recorte de etiquetas
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
