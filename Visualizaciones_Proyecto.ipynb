```python
!pip install plotly pandas

import plotly.express as px
import pandas as pd

# Datos proporcionados
artistas = ["Dos Hielos - King Savagge", "XQ TAN SOLA? – Nickoog Clk", "X ESO BB - Jere Klein", "Llámame Bebé  - Pailita", "Donde? Jere Klein"]
semanas = ["Reproducciones primera semana", "Reproducciones segunda semana", "Reproducciones tercera semana"]

reproducciones = [
    [1997331, 1966104, 3237932, 3960736, 746892],
    [2061894, 2554367, 2956073, 3681577, 1187723],
    [2177013, 2634225, 3465663, 3232945, 1742747]
]

# Crear un DataFrame con los datos
df = pd.DataFrame(reproducciones, columns=artistas, index=semanas)

# Crear el gráfico interactivo con Plotly
fig = px.line(df, x=df.index, y=artistas, markers=True, title='Reproducciones por semana - Artistas chilenos',
              labels={'value': 'Reproducciones', 'variable': 'Artista'},
              template='plotly_dark')

# Agregar información emergente al pasar el cursor sobre los puntos
fig.update_traces(hovertemplate='%{y:,} reproducciones')

# Mostrar el gráfico
fig.show()


```

  **Explicación**: Este es un gráfico de líneas que representa las canciones de artistas chilenos más escuchadas en nuestro país durante 2023. La visualización cuenta con un seguimiento de las tres semanas posteriores al estreno de cada canción. El gráfico cuenta con cinco líneas. Cada una de ellas representa una canción. En el eje X tenemos cada semana (primera, segunda y tercera) y en el eje Y tenemos las cantidades de reproducciones. Los puntos se refieren a cuántas escuchas llegó la canción en cada semana. En tres de ellas (Donde?, XQ TAN SOLA? y 2 Hielos) en cuanto pasan las semanas, las reproducciones van aumentando, mientras que en las otras dos (Llámame Bebé y X ESO BB) van decreciendo desde la primera semana.
**Mensaje a comunicar**: Lo que se quiere comunicar con este gráfico es el apoyo que reciben los artistas chilenos apenas salen las canciones. En la primera semana todas tienen más de un millón (salvo Donde? que salió por primera vez dentro de un álbum, lo que hace más difícil su escucha), incluso casi dos millones y otras que llegan a los tres. Lo importante de esto es la constancia y el aumento de estas cifras. A medida que pasa el tiempo, las reproducciones van subiendo y si bajan son en cantidades muy pequeñas. Entonces lo que queremos transmitir con este gráfico es el gran colchón que tiene Chile (y América Latina) para sus artistas, partiendo en las primeras semanas con gran recibimiento que luego se traduce en un salto de catapulta al resto del mundo.

```python
import pandas as pd
import altair as alt

# Crear un DataFrame con los datos proporcionados
data = {
    'Canciones con mayor Streams 2023': ["Flowers - Miley Cyrus", "BZRP Shakira y Bizarrap", "Ella baila sola - Eslabón Armado y Peso Pluma",
                                        "TQG - Shakira y Karol G", "Boy's a liar Pt. 2 - PinkPantheress e Ice Spice",
                                        "Seven - Jung Kook y Latto", "La bebé (remix) - Yng Lvcas y Peso Pluma",
                                        "Cupid (twin ver) - Fifty fifty", "Un x100to - Grupo Frontera y Bad Bunny", "Daylight - David Kushner"],
    'Streams 2023 (en millones)': [1611403194, 834949142, 996003088, 819829846, 729950285, 1142338508, 817820824, 711619265, 769557440, 740759015],
    'Estreno': ["enero", "enero", "marzo", "febrero", "febrero", "julio", "marzo", "febrero", "abril", "abril"]
}

df = pd.DataFrame(data)

# Definir el orden personalizado de los meses
order = ["enero", "febrero", "marzo", "abril", "julio"]

# Crear un nuevo mapeo de colores intercalados
color_mapping_intercalado = {
    "enero": "#000000",   # Negro
    "febrero": "#4285F4", # Azul
    "marzo": "#1DB954",    # Verde
    "abril": "#FFD700",    # Amarillo
    "julio": "#808080"    # Repetir Negro para completar el ciclo
}
df['Color'] = df['Estreno'].map(color_mapping_intercalado)

# Ordenar el DataFrame por Streams de manera descendente
df = df.sort_values(by='Streams 2023 (en millones)', ascending=False)

# Crear un gráfico de barras horizontales con Altair usando colores intercalados y leyenda
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Streams 2023 (en millones):Q', axis=alt.Axis(format=".3s")),
    y=alt.Y('Canciones con mayor Streams 2023:N', sort=None),  # Dejar que Altair use el orden del DataFrame
    color=alt.Color('Estreno:N', scale=alt.Scale(domain=order, range=list(color_mapping_intercalado.values())), legend=alt.Legend(title='Mes de Estreno', orient='left')),  # Añadir leyenda
    tooltip=['Canciones con mayor Streams 2023', 'Streams 2023 (en millones):Q', 'Estreno']
).properties(width=600, height=400, title='Canciones con Mayor Streams en 2023')

# Añadir texto con las reproducciones exactas al lado de cada barra
text = chart.mark_text(
    align='left',
    baseline='middle',
    dx=3,  # Desplazamiento horizontal para evitar solapamiento
).encode(
    text='Streams 2023 (en millones):Q'
)

# Combinar el gráfico de barras y el texto
chart_with_text = (chart + text)

# Mostrar el gráfico
chart_with_text


# Mostrar el gráfico
chart_with_text

```

**Explicación**: El último gráfico de nuestra webstory corresponde a una graficación, a nivel mundial, de las canciones con más reproducciones en 2023, siendo 5 de ellas, equivalente a la mitad, provenientes de artistas de América Latina.
**Mensaje a comunicar**: Esto indica que del total de canciones con más reproducciones a nivel mundial durante el año, la mitad son latinoamericanas. Esto deja en evidencia el crecimiento de la influencia latina, y cómo ha logrado exponer en distintas partes de otros continentes. Se puede analizar como en latinoamerica ha habido un proceso de enriqueicmiento de la música urbana, cuidandola e impulsándola a dar un salto hacia los rankings globales.
