```python
import matplotlib.pyplot as plt

# Datos
canciones = ["La Morocha - Lucka Ra / BM", "Lollipop - Darell", "Fruto - Bizarrap / Milo j", "Una Foto - Mesita", "Linda (Remix) - Marka / DJ Tao / Lauty Gram / Migrantes / Peipper"]
esta_semana = [3713785, 3666880, 3147017, 3015834, 2964808]
semana_pasada = [3988746, 3594490, 3807991, 3108760, 2915357]
semana_antepasada = [4302674, 1722824, 3960414, 3353308, 2441215]

# Crear el gráfico de líneas con ejes estándar
plt.figure(figsize=(10, 6))

for i, cancion in enumerate(canciones):
    plt.plot(["Semana Antepasada", "Semana Pasada", "Esta Semana"],
             [semana_antepasada[i], semana_pasada[i], esta_semana[i]],
             label=cancion, marker='o')

# Personalizar el gráfico
plt.xlabel('Semanas')
plt.ylabel('Reproducciones (en millones)')
plt.title('Reproducciones de Canciones por Semana')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
canciones = ["La Morocha - Lucka Ra / BM", "Lollipop - Darell", "Fruto - Bizarrap / Milo j", "Una Foto - Mesita", "Linda (Remix) - Marka / DJ Tao / Lauty Gram / Migrantes / Peipper"]
esta_semana = [3713785, 3666880, 3147017, 3015834, 2964808]
semana_pasada = [3988746, 3594490, 3807991, 3108760, 2915357]
semana_antepasada = [4302674, 1722824, 3960414, 3353308, 2441215]

# Crear arrays para la posición de las barras
bar_height = 0.2
index = np.arange(len(canciones))

# Crear el gráfico de barras horizontales
plt.figure(figsize=(10, 6))

bar1 = plt.barh(index - bar_height, esta_semana, bar_height, label='Esta Semana')
bar2 = plt.barh(index, semana_pasada, bar_height, label='Semana Pasada')
bar3 = plt.barh(index + bar_height, semana_antepasada, bar_height, label='Semana Antepasada')

# Personalizar el gráfico
plt.ylabel('Canciones')
plt.xlabel('Reproducciones')
plt.title('Reproducciones de Canciones por Semana')
plt.yticks(index, canciones)
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Nombres actualizados de las canciones
canciones = ["La Morocha - Lucka Ra / BM", "Lollipop - Darell", "Fruto - Bizarrap / Milo j", "Una Foto - Mesita", "Linda (Remix) - Marka / DJ Tao / Lauty Gram / Migrantes / Peipper"]
esta_semana = [3713785, 3666880, 3147017, 3015834, 2964808]

# Crear un array de índices para la ubicación de las barras
ind = np.arange(len(canciones))

# Ajustar el ancho de la figura y el ancho de las barras
fig = plt.figure(figsize=(10, 6))
width = 0.7

# Crear el gráfico de barras
bars = plt.bar(ind, esta_semana, width, color='blue')

# Etiquetas, título y leyenda
plt.xlabel('Canciones')
plt.ylabel('Número de reproducciones')
plt.title('Top 5 Canciones en Argentina - Esta Semana')
plt.xticks(ind, canciones, rotation=15, ha="right")

# Ajustar el límite superior del eje y para dejar espacio adicional
plt.ylim(0, max(esta_semana) * 1.1)

# Formatear el eje y para quitar los puntos decimales
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: format(int(x), ',')))

# Añadir las cifras exactas arriba de las barras
for bar, count in zip(bars, esta_semana):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.02 * max(esta_semana), f'{count:,}', ha='center', va='bottom')

# Ajustar el diseño para evitar recorte de etiquetas
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
